#Informações

Aqui está uma breve explicação de cada um:

1. **Cotação (Cotação)**: Preço atual da ação no mercado.

2. **P/L (Preço/Lucro)**: Também conhecido como P/E (Price/Earnings), é o preço da ação dividido pelo lucro por ação. Indica quantos anos seriam necessários para recuperar o investimento inicial pelo lucro gerado pela empresa.

3. **P/VP (Preço/Valor Patrimonial)**: Relação entre o preço da ação e o valor patrimonial por ação. Indica quanto o mercado está disposto a pagar pelo patrimônio líquido da empresa.

4. **PSR (Preço/Serviços Prestados)**: Relação entre o preço da ação e a receita líquida por ação. Mostra quanto o mercado está disposto a pagar pela receita gerada pela empresa.

5. **Dividend Yield (Div.Yield)**: Indica o retorno em dividendos que um acionista pode esperar receber em relação ao preço atual da ação.

6. **P/Ativo (P/Ativo)**: Relação entre o preço da ação e o valor do ativo total por ação. Indica quanto o mercado está disposto a pagar pelo ativo da empresa.

7. **P/Cap.Giro (Preço/Capital de Giro)**: Relação entre o preço da ação e o capital de giro por ação. Mostra quanto o mercado está disposto a pagar pelo capital de giro da empresa.

8. **P/EBIT (Preço/EBIT)**: Relação entre o preço da ação e o lucro antes dos juros e impostos por ação.

9. **P/Ativ Circ.Liq (Preço/Ativo Circulante Líquido)**: Relação entre o preço da ação e o ativo circulante líquido por ação. Indica quanto o mercado está disposto a pagar pelo ativo circulante líquido da empresa.

10. **EV/EBIT (Enterprise Value/EBIT)**: Relação entre o valor da empresa e o lucro antes dos juros e impostos. Mostra quanto o mercado está disposto a pagar pelo lucro operacional da empresa.

11. **EV/EBITDA (Enterprise Value/EBITDA)**: Relação entre o valor da empresa e o lucro antes dos juros, impostos, depreciação e amortização. Mostra quanto o mercado está disposto a pagar pelo lucro operacional mais as despesas não monetárias.

12. **Margem EBIT (Mrg Ebit)**: Margem operacional da empresa, calculada como o lucro antes dos juros e impostos dividido pela receita líquida.

13. **Margem Líquida (Mrg. Líq.)**: Margem líquida da empresa, calculada como o lucro líquido dividido pela receita líquida.

14. **Liquidez Corrente (Liq. Corr.)**: Relação entre o ativo circulante e o passivo circulante. Indica a capacidade da empresa de pagar suas obrigações de curto prazo.

15. **ROIC (Return on Invested Capital)**: Retorno sobre o capital investido, calculado como o lucro líquido dividido pelo capital investido.

16. **ROE (Return on Equity)**: Retorno sobre o patrimônio líquido, calculado como o lucro líquido dividido pelo patrimônio líquido.

17. **Liquidez 2 Meses (Liq.2meses)**: Relação entre o volume médio diário negociado nos últimos 2 meses e o total de ações emitidas pela empresa.

18. **Patrimônio Líquido (Patrim. Líq)**: Valor contábil do patrimônio líquido da empresa.

19. **Dívida Bruta/Patrimônio (Dív.Brut/ Patrim.)**: Relação entre a dívida bruta e o patrimônio líquido da empresa.

20. **Crescimento Receita 5 Anos (Cresc. Rec.5a)**: Taxa de crescimento média anual dos últimos 5 anos da receita da empresa.

# Coleta de dados

```python
# Buscando as informações contidas no API
df_trabalho = fundamentus.get_resultado_raw()
df_trabalho.head()
```

```python
# Quantidade de linhas e colunas
print(f'Linhas: {df_trabalho.shape[0]} e colunas: {df_trabalho.shape[1]}')
```

```python
# Como o objetivo neste momento é buscar informações das empresas e não de cada ação - devemos tratar as duplicidades - exemplo
exemplos_de_acoes_duplicadas = df_trabalho[df_trabalho.index.str[:4] == 'PETR']
exemplos_de_acoes_duplicadas
```

```python
# Verificando duplicidade de informações
df_trabalho['Empresa'] = df_trabalho.index.str[:4]
duplicatas = df_trabalho.duplicated(subset='Empresa', keep=False).sum()
duplicatas
```

```python
# Verificando duplicidade de informações
df_trabalho['Empresa'] = df_trabalho.index.str[:4]
df_trabalho['Ticker'] = df_trabalho.index

# Adicionar ".SA" ao final de cada valor na coluna 'Ticker'
df_trabalho['Ticker'] =  df_trabalho.index + '.SA'

# Remover duplicidades
df_trabalho = df_trabalho.drop_duplicates(subset='Empresa')
df_trabalho = df_trabalho.drop(columns=['Empresa'])

# Criando novo indice por empresa
#df_trabalho = df_trabalho.reset_index(drop=True)
#df_trabalho = df_trabalho.set_index('Empresa')

# Quantidade de linhas e colunas
print(f'Linhas atuais: {df_trabalho.shape[0]}')

df_trabalho.head()
```

```python
from datetime import datetime, timedelta

def is_traded_recently(acao):
    try:
        # Trazer os dados históricos de cotação para aquela ação
        stock = yf.Ticker(acao)
        historical_data = stock.history(period='1mo')

        # Trazer a cotação mais recente e obter a data correspondente
        last_trading_day = historical_data.index[-1].date()

        # Verificar se o dia de negociação mais recente está dentro de uma janela de dias
        today = datetime.now().date()
        window_start = today - timedelta(days=5)

        return window_start <= last_trading_day <= today
    except Exception:

        return False

# Aplicar a função a cada linha do DataFrame com tqdm
df_trabalho['Negociada'] = df_trabalho['Ticker'].apply(is_traded_recently)

df_trabalho.tail()
```

```python
# Agrupando feito pela informação de negociada ou não
contagem = df_trabalho.groupby('Negociada').size().reset_index(name='Quantidade')

# Mostrando os grupos separadamente
for index, row in contagem.iterrows():
    print(f"Negociada: {row['Negociada']}, Quantidade: {row['Quantidade']}")
```

```python
# Tratando a informação de crescimento

df_trabalho['Cresc. Rec.5a'] = (df_trabalho['Cresc. Rec.5a'] * 100 )
```

# Apresentando algumas informações

```python
#
maximo = (df_trabalho['Cresc. Rec.5a'].max()).round(2)
minimo = (df_trabalho['Cresc. Rec.5a'].min()).round(2)
print(f"Os valores de crescimento estão variando de {minimo}.round(2) a '{maximo}")
```

```python
# Filtrando os dados
criérios_selecao = [0,0.5,50]
tickers_selecionados = df_trabalho[ (df_trabalho['Negociada'] == True) &
                                    (df_trabalho['Patrim. Líq'] > criérios_selecao[0]) &
                                    (df_trabalho['Dív.Brut/ Patrim.'] < criérios_selecao[1]) &
                                    (df_trabalho['Cresc. Rec.5a'] >= criérios_selecao[2])]

tickers_por_cresc_e_patrimonio = tickers_selecionados.sort_values('Dív.Brut/ Patrim.', ascending=True)
tickers_por_cresc_e_patrimonio
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime

# Obter a data e hora atual
data_e_hora = datetime.now().strftime("%d/%m/%Y %H:%M:%S")

# Ordenar os dados
tickers_por_cresc_e_patrimonio = tickers_selecionados.sort_values('Cresc. Rec.5a', ascending=False)

# Paleta de cores
colors = sns.color_palette("viridis", len(tickers_por_cresc_e_patrimonio))

# Plotar o gráfico de barras
plt.figure(figsize=(12, 6))
sns.barplot(x='Ticker', y='Cresc. Rec.5a', data=tickers_por_cresc_e_patrimonio, palette=colors)

# Adicionar os valores da barra como legenda
for index, row in tickers_por_cresc_e_patrimonio.iterrows():
    plt.annotate(f'{row["Cresc. Rec.5a"]:.2f}', (index, row["Cresc. Rec.5a"]), ha='center', va='center')

plt.xlabel('Ticker', fontsize=14)
plt.ylabel('Cresc. Rec.5a', fontsize=14)
plt.suptitle('Crescscimento Recorrente (5a) das Empresas com ações em negociação na B3 selecionadas', fontsize=14)
plt.title(f'Critérios: Patrim. Líq maior que {criérios_selecao[0]}, Dív.Brut/ Patrim. menor que {criérios_selecao[1]} e Cresc. Rec.5a maior ou igual a {criérios_selecao[2]}', fontsize=12, y=1.05, ha='center')
plt.xticks(rotation=45, ha='right', fontsize=10)
plt.yticks(fontsize=10)
plt.figtext(0.99, 0.01, f'Gráfico gerado em: {data_e_hora}', horizontalalignment='right')
plt.tight_layout()
plt.show()
```
